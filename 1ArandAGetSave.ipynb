# Stores Adjacency matrices when random (no rewiring) and after 4000 rewirings

They are stored in a dictionary

```python
vertices = 100
edges = int(np.round(2 * np.log(vertices) * (vertices - 1), decimals=0))
taus = {}
taus['normal'] = [4 ,4.1 ,4.2, 4.3, 4.4]
taus['lognormal'] = [5.6, 5.7, 5.8 ,5.9, 6]
pRand = [0.2]

rewirings = 4000
repetitions = 1000
```

```python
directorySave ='data/ArandA/1000iterationsHist/'
weightDist = ['normal','lognormal']
dictArandA = {}
for wD in weightDist:
    filePathSave = directorySave + 'ArandA_tauTransProx_'+wD+'_p'+str(pRand[0])+'_rewir'+str(rewirings)+'.pckl'
    print(filePathSave)
    dictArandA[wD] = ce.getArandAManyIterations(vertices, edges, taus[wD], pRand, rewirings, wD, repetitions)
    hf.saveVarSimple(dictArandA[wD], filePathSave)
```
