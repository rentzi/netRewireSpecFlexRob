## Functions

```python
def scatterPlotQRandQRewNormalVsLognormal(Q,p,rewirings = 4000,filePath = False):

    weightDist = ['normal','lognormal']
    colors = [ 'orange', 'green']
    alphas = [0.2,0.2]
    markers = ['s','v']

    plt.rcParams['figure.figsize'] = [5, 5]

    fig = plt.figure()
    ax = fig.add_subplot(111)
    ## This is the key.  elements with zorder < 1 will be rasterized
    ax.set_rasterization_zorder(1) 
    for ind,wD in enumerate(weightDist):
        ax.scatter(Q[(wD,p)][0],Q[(wD,p)][rewirings],alpha = alphas[ind],color = colors[ind], marker = markers[ind], zorder=0)
        plt.xlim(0.13, 0.41)
        plt.ylim(0, 0.85) 
    
    if filePath is not False:
        directory = os.path.dirname(filePath)
        if not os.path.exists(directory):  # makes the directory if it does not exist
            os.makedirs(directory)
        fig.savefig(filePath, rasterized=True, dpi=1200)   
```
